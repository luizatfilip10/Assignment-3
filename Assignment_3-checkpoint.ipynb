### Import Pandas, Requests and BeautifulSoup

```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
import re
```

### Web Scraping

```python
metadata = pd.read_csv('metadataA3.csv')

descriptions = []
price = []
availability = []
ratings = []
rating_map = {"One": 1, "Two": 2, "Three": 3, "Four": 4, "Five": 5}

for url in metadata['URL']:
    response = requests.get(url)
    document = BeautifulSoup(response.text, "html.parser")

    #Product Description
    description_content = document.find("div", id = "product_description")
    if description_content:
        desc_p = description_content.find_next_sibling("p") #since the "product description" header and the actual description are two different sections
        if desc_p:
            descriptions.append(desc_p.get_text(strip=True))
        else:
            descriptions.append("")
    else:
        descriptions.append("")

    #Price
    price = document.find("p", class_ = "price_color").text

    #Availability
    availability = document.find("p", class_ = "instock availability").text.strip()

    #Ratings and converting from text to number

    rating_tag = document.find("p", class_="star-rating")
    if rating_tag:
        rating_text = rating_tag.get("class")[1] 
        ratings.append(rating_map.get(rating_text,0))
    else:
        ratings.append(0)


metadata['Description'] = descriptions
metadata['Price'] = price
metadata['Availability'] = availability
metadata['Rating'] = ratings
metadata
```

```python
metadata.to_csv("scraped_data_A3.csv")
```
